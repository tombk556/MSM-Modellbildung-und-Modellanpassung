```python
from sympy import symbols, Function, Eq, diff, simplify
```

## Aufgabe 1)

### b)

```python
### f(x) definieren
y, r, x, k = symbols("y(x) r x k")
fx = Eq(y, r*x*(1 - x/k) - ((x**2)/ (1 + x**2)))

fx
```

```python
### Fixpunkt berechnen x*=0, also f(0) sollte 0 ergeben...

fx.subs(x, 0)
```

```python
### Stabilitätsanalyse

# 1. Schritt: ableiten...

derivative_rhs = diff(fx.rhs, x)

y_prime = symbols("y'(x)", real=True)
fx_der = Eq(y_prime, derivative_rhs)

# 2. Schritt: f'(x = 0)
print("für x = 0")
fx_der.subs(x, 0)


### -> der Fixpunkt is instabil
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f_{k,r}(x) for various k and r
def f_kr(x, r, k):
    return r * x * (1 - x / k) - x**2 / (1 + x**2)

# Define values of k and r for different cases
k_values = [1, 2, 3, 4]
r_values = [0.5, 1, 1.5, 2]

# Generate x values for plotting
x = np.linspace(0, 5, 400)

# Plot f_{k,r}(x) for various k and r
plt.figure(figsize=(10, 6))
for k in k_values:
    for r in r_values:
        y = f_kr(x, r, k)
        plt.plot(x, y, label=f'k={k}, r={r}')

plt.axhline(0, color='black',linewidth=1, linestyle='--')
plt.title('Plots of f_{k,r}(x) for Various k and r Values')
plt.xlabel('x')
plt.ylabel('f_{k,r}(x)')
plt.legend()
plt.grid(True)
plt.show()
```
