```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

t = np.array([10, 20, 40, 60, 90, 120])  # Zeit in Sekunden
T = np.array([275, 212, 185, 123, 74, 62])  # Temperatur in 째C

def temperature_model(t, a, T0, k):
    return a + (T0 - a) * np.exp(-k * t)

popt, pcov = curve_fit(temperature_model, t, T)
a, T0, k = popt

print("Gesch채tzte Parameter:")
print(f"Umgebungstemperatur (a): {a:.2f} 째C")
print(f"Anfangstemperatur (T0): {T0:.2f} 째C")
print(f"Proportionalit채tskonstante (k): {k:.4f}")

t_fit = np.linspace(0, 130, 500)
T_fit = temperature_model(t_fit, a, T0, k)

plt.scatter(t, T, label="Messdaten", color="red")
plt.plot(t_fit, T_fit, label="Angepasste Kurve", linestyle="--")
plt.xlabel("Zeit (s)")
plt.ylabel("Temperatur (째C)")
plt.title("Temperaturverlauf der Bremsscheibe")
plt.legend()
plt.show()
```
